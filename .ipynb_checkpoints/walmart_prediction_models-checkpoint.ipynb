# Walmart Inventory Prediction Models
## Complete Guide: Traditional ML + Deep Learning

### Predictions We'll Build:
1. **Inventory Level Prediction** (Regression - Deep Learning)
2. **Reorder Classification** (Binary Classification - Deep Learning)
3. **Demand Forecasting** (Quantity Sold - Traditional ML vs DL)
4. **Revenue Prediction** (Regression - Comparison)

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Traditional ML Models
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.ensemble import GradientBoostingRegressor

# Deep Learning
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Load Data
df = pd.read_csv('Walmart-Inventory-Dataset.csv')
print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
df.head()
```

## 1. Data Preprocessing

```python
# Create a copy for preprocessing
data = df.copy()

# Convert transaction_date to datetime and extract features
data['transaction_date'] = pd.to_datetime(data['transaction_date'])
data['year'] = data['transaction_date'].dt.year
data['month'] = data['transaction_date'].dt.month
data['day'] = data['transaction_date'].dt.day
data['day_of_week'] = data['transaction_date'].dt.dayofweek
data['quarter'] = data['transaction_date'].dt.quarter

# Create additional features
data['revenue'] = data['quantity_sold'] * data['unit_price']
data['inventory_to_reorder_ratio'] = data['inventory_level'] / data['reorder_point']
data['needs_reorder'] = (data['inventory_level'] <= data['reorder_point']).astype(int)
data['stockout_risk'] = (data['inventory_level'] < data['reorder_point'] * 0.5).astype(int)

print("✓ Date features extracted")
print("✓ Revenue calculated")
print("✓ Target variables created")
print(f"\nNeeds Reorder Distribution:\n{data['needs_reorder'].value_counts()}")
```

```python
# Check for missing values
print("Missing Values:")
print(data.isnull().sum())

# Info about data types
data.info()
```

## 2. Feature Engineering & Encoding

## 3. MODEL 1: Inventory Level Prediction (Deep Learning)
### Regression Problem - Predict exact inventory levels

```python
# Select features for inventory prediction
inventory_features = [
    'quantity_sold', 'unit_price', 'reorder_point', 'reorder_quantity',
    'supplier_lead_time', 'customer_age', 'customer_income',
    'product_name_encoded', 'category_encoded', 'store_location_encoded',
    'customer_gender_encoded', 'customer_loyalty_level_encoded',
    'payment_method_encoded', 'discount_applied_encoded',
    'season_encoded', 'promotional_campaign_encoded', 'weather_condition_encoded',
    'month', 'day_of_week', 'quarter'
]

X_inventory = data[inventory_features]
y_inventory = data['inventory_level']

# Train-test split
X_train_inv, X_test_inv, y_train_inv, y_test_inv = train_test_split(
    X_inventory, y_inventory, test_size=0.2, random_state=42
)

# Scale features (IMPORTANT for Deep Learning!)
scaler_inv = StandardScaler()
X_train_inv_scaled = scaler_inv.fit_transform(X_train_inv)
X_test_inv_scaled = scaler_inv.transform(X_test_inv)

print(f"Training set: {X_train_inv_scaled.shape}")
print(f"Test set: {X_test_inv_scaled.shape}")
```

```python
# Build Deep Learning Model for Inventory Prediction
def build_inventory_model(input_dim):
    model = models.Sequential([
        # Input layer
        layers.Dense(128, activation='relu', input_shape=(input_dim,)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        # Hidden layers
        layers.Dense(64, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.2),
        
        layers.Dense(16, activation='relu'),
        
        # Output layer (regression - no activation)
        layers.Dense(1)
    ])
    
    # Compile model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )
    
    return model

# Create model
inventory_model = build_inventory_model(X_train_inv_scaled.shape[1])

# Model summary
inventory_model.summary()
```

```python
# Define callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=0.00001,
    verbose=1
)

# Train the model
history_inv = inventory_model.fit(
    X_train_inv_scaled, y_train_inv,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

print("\n✓ Inventory Model Training Complete!")
```

```python
# Plot training history
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss
axes[0].plot(history_inv.history['loss'], label='Training Loss')
axes[0].plot(history_inv.history['val_loss'], label='Validation Loss')
axes[0].set_title('Inventory Model - Loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss (MSE)')
axes[0].legend()
axes[0].grid(True)

# MAE
axes[1].plot(history_inv.history['mae'], label='Training MAE')
axes[1].plot(history_inv.history['val_mae'], label='Validation MAE')
axes[1].set_title('Inventory Model - Mean Absolute Error')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MAE')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Evaluate Inventory Model
y_pred_inv = inventory_model.predict(X_test_inv_scaled).flatten()

# Calculate metrics
mse_inv = mean_squared_error(y_test_inv, y_pred_inv)
rmse_inv = np.sqrt(mse_inv)
mae_inv = mean_absolute_error(y_test_inv, y_pred_inv)
r2_inv = r2_score(y_test_inv, y_pred_inv)

print("="*50)
print("INVENTORY LEVEL PREDICTION - DEEP LEARNING MODEL")
print("="*50)
print(f"Mean Squared Error (MSE): {mse_inv:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_inv:.2f}")
print(f"Mean Absolute Error (MAE): {mae_inv:.2f}")
print(f"R² Score: {r2_inv:.4f}")
print("="*50)
```

```python
# Visualize predictions
plt.figure(figsize=(12, 5))

# Scatter plot
plt.subplot(1, 2, 1)
plt.scatter(y_test_inv, y_pred_inv, alpha=0.5)
plt.plot([y_test_inv.min(), y_test_inv.max()], 
         [y_test_inv.min(), y_test_inv.max()], 'r--', lw=2)
plt.xlabel('Actual Inventory Level')
plt.ylabel('Predicted Inventory Level')
plt.title('Actual vs Predicted - Inventory Level')
plt.grid(True)

# Residuals
plt.subplot(1, 2, 2)
residuals = y_test_inv - y_pred_inv
plt.scatter(y_pred_inv, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--', lw=2)
plt.xlabel('Predicted Inventory Level')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 4. MODEL 2: Reorder Classification (Deep Learning)
### Binary Classification - Predict if reorder is needed

```python
# Prepare data for reorder classification
reorder_features = [
    'quantity_sold', 'unit_price', 'inventory_level', 'reorder_point',
    'reorder_quantity', 'supplier_lead_time', 'customer_age', 'customer_income',
    'product_name_encoded', 'category_encoded', 'store_location_encoded',
    'customer_loyalty_level_encoded', 'season_encoded',
    'month', 'day_of_week', 'quarter'
]

X_reorder = data[reorder_features]
y_reorder = data['needs_reorder']

# Train-test split
X_train_reo, X_test_reo, y_train_reo, y_test_reo = train_test_split(
    X_reorder, y_reorder, test_size=0.2, random_state=42, stratify=y_reorder
)

# Scale features
scaler_reo = StandardScaler()
X_train_reo_scaled = scaler_reo.fit_transform(X_train_reo)
X_test_reo_scaled = scaler_reo.transform(X_test_reo)

print(f"Training set: {X_train_reo_scaled.shape}")
print(f"Class distribution in training:")
print(y_train_reo.value_counts())
```

```python
# Build Classification Model
def build_reorder_model(input_dim):
    model = models.Sequential([
        # Input layer
        layers.Dense(64, activation='relu', input_shape=(input_dim,)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        # Hidden layers
        layers.Dense(32, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        
        layers.Dense(16, activation='relu'),
        layers.Dropout(0.2),
        
        # Output layer (binary classification - sigmoid activation)
        layers.Dense(1, activation='sigmoid')
    ])
    
    # Compile model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', keras.metrics.Precision(), keras.metrics.Recall()]
    )
    
    return model

# Create model
reorder_model = build_reorder_model(X_train_reo_scaled.shape[1])

# Model summary
reorder_model.summary()
```

```python
# Train the classification model
history_reo = reorder_model.fit(
    X_train_reo_scaled, y_train_reo,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

print("\n✓ Reorder Classification Model Training Complete!")
```

```python
# Plot training history
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss
axes[0].plot(history_reo.history['loss'], label='Training Loss')
axes[0].plot(history_reo.history['val_loss'], label='Validation Loss')
axes[0].set_title('Reorder Model - Loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Binary Crossentropy Loss')
axes[0].legend()
axes[0].grid(True)

# Accuracy
axes[1].plot(history_reo.history['accuracy'], label='Training Accuracy')
axes[1].plot(history_reo.history['val_accuracy'], label='Validation Accuracy')
axes[1].set_title('Reorder Model - Accuracy')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Evaluate Reorder Classification Model
y_pred_reo_prob = reorder_model.predict(X_test_reo_scaled).flatten()
y_pred_reo = (y_pred_reo_prob > 0.5).astype(int)

# Calculate metrics
accuracy = accuracy_score(y_test_reo, y_pred_reo)

print("="*50)
print("REORDER CLASSIFICATION - DEEP LEARNING MODEL")
print("="*50)
print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test_reo, y_pred_reo, 
                          target_names=['No Reorder', 'Needs Reorder']))
print("="*50)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test_reo, y_pred_reo)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Reorder', 'Needs Reorder'],
            yticklabels=['No Reorder', 'Needs Reorder'])
plt.title('Confusion Matrix - Reorder Classification')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

## 5. MODEL 3: Demand Forecasting (Quantity Sold)
### Compare Traditional ML vs Deep Learning

```python
# Prepare data for demand forecasting
demand_features = [
    'unit_price', 'inventory_level', 'customer_age', 'customer_income',
    'product_name_encoded', 'category_encoded', 'store_location_encoded',
    'customer_loyalty_level_encoded', 'payment_method_encoded',
    'discount_applied_encoded', 'season_encoded', 'promotional_campaign_encoded',
    'weather_condition_encoded', 'month', 'day_of_week', 'quarter'
]

X_demand = data[demand_features]
y_demand = data['quantity_sold']

# Train-test split
X_train_dem, X_test_dem, y_train_dem, y_test_dem = train_test_split(
    X_demand, y_demand, test_size=0.2, random_state=42
)

# Scale for deep learning
scaler_dem = StandardScaler()
X_train_dem_scaled = scaler_dem.fit_transform(X_train_dem)
X_test_dem_scaled = scaler_dem.transform(X_test_dem)

print(f"Training set: {X_train_dem.shape}")
print(f"Test set: {X_test_dem.shape}")
```

```python
# A. Traditional ML - Random Forest
print("Training Random Forest...")
rf_demand = RandomForestRegressor(n_estimators=100, max_depth=20, 
                                 random_state=42, n_jobs=-1)
rf_demand.fit(X_train_dem, y_train_dem)
y_pred_rf = rf_demand.predict(X_test_dem)

# Random Forest metrics
mse_rf = mean_squared_error(y_test_dem, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
mae_rf = mean_absolute_error(y_test_dem, y_pred_rf)
r2_rf = r2_score(y_test_dem, y_pred_rf)

print("\n✓ Random Forest trained!")
```

```python
# B. Deep Learning Model for Demand
def build_demand_model(input_dim):
    model = models.Sequential([
        layers.Dense(128, activation='relu', input_shape=(input_dim,)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(64, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.2),
        
        layers.Dense(1)
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )
    
    return model

# Create and train
demand_model = build_demand_model(X_train_dem_scaled.shape[1])

history_dem = demand_model.fit(
    X_train_dem_scaled, y_train_dem,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping, reduce_lr],
    verbose=0
)

# DL predictions
y_pred_dl = demand_model.predict(X_test_dem_scaled, verbose=0).flatten()

# DL metrics
mse_dl = mean_squared_error(y_test_dem, y_pred_dl)
rmse_dl = np.sqrt(mse_dl)
mae_dl = mean_absolute_error(y_test_dem, y_pred_dl)
r2_dl = r2_score(y_test_dem, y_pred_dl)

print("✓ Deep Learning model trained!")
```

```python
# Compare Random Forest vs Deep Learning
comparison_df = pd.DataFrame({
    'Metric': ['RMSE', 'MAE', 'R² Score'],
    'Random Forest': [rmse_rf, mae_rf, r2_rf],
    'Deep Learning': [rmse_dl, mae_dl, r2_dl]
})

print("="*60)
print("DEMAND FORECASTING - MODEL COMPARISON")
print("="*60)
print(comparison_df.to_string(index=False))
print("="*60)

# Determine winner
if r2_rf > r2_dl:
    print(f"\n🏆 Winner: Random Forest (R² = {r2_rf:.4f})")
else:
    print(f"\n🏆 Winner: Deep Learning (R² = {r2_dl:.4f})")
```

```python
# Visualize comparison
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Random Forest
axes[0].scatter(y_test_dem, y_pred_rf, alpha=0.5, color='green')
axes[0].plot([y_test_dem.min(), y_test_dem.max()], 
             [y_test_dem.min(), y_test_dem.max()], 'r--', lw=2)
axes[0].set_xlabel('Actual Quantity Sold')
axes[0].set_ylabel('Predicted Quantity Sold')
axes[0].set_title(f'Random Forest (R² = {r2_rf:.4f})')
axes[0].grid(True)

# Deep Learning
axes[1].scatter(y_test_dem, y_pred_dl, alpha=0.5, color='blue')
axes[1].plot([y_test_dem.min(), y_test_dem.max()], 
             [y_test_dem.min(), y_test_dem.max()], 'r--', lw=2)
axes[1].set_xlabel('Actual Quantity Sold')
axes[1].set_ylabel('Predicted Quantity Sold')
axes[1].set_title(f'Deep Learning (R² = {r2_dl:.4f})')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
