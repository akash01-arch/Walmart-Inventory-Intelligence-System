```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Walmart-Inventory-Dataset.csv')
```

***What is the total revenue generated?***

```python
df['Revenue'] = df['quantity_sold'] * df['unit_price']
```

```python
total_revenue = df['Revenue'].sum()
print(f'The Total Revenue Generatede: {total_revenue}')
```

***Which categories generate the highest revenue?***

```python
category_revenue = df.groupby('category')['Revenue'].sum().sort_values(ascending=False)
print("Revenue by Category:\n", category_revenue)
```

***Which products sell more but earn less revenue?***

```python
df['category'].max()
```

```python
Target_products = df.sort_values(by=['quantity_sold', 'Revenue'], ascending=[False,True])
# Target_products
```

***Store-wise revenue comparison.***

```python
store_revenue = df.groupby('store_location')['Revenue'].sum().sort_values(ascending=False)
print("Revenue by Store:\n", store_revenue)
```

```python
import matplotlib.ticker as ticker
ax = store_revenue.plot(kind='bar', color='skyblue')

formatter = ticker.FuncFormatter(lambda x, p: f'${x*1e-6:1.1f}M')
ax.yaxis.set_major_formatter(formatter)

plt.title('Total Revenue by Store Location')
plt.ylabel('Revenue (USD)')
plt.show()
```

***Monthly sales trends â€“ any seasonality?***

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['YearMonth'] = df['transaction_date'].dt.to_period('M')
```

```python
monthly_revenue = df.groupby('YearMonth')['Revenue'].sum()
print(monthly_revenue)
```

```python
ax = monthly_revenue.plot(marker='o', linestyle='-', color='blue')

ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, p: f'${x:,.0f}')
)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Revenue (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
stats = df.groupby('category').agg({
    'quantity_sold': 'sum',
    'Revenue': 'sum'
}).reset_index()

high_vol_low_rev = stats[
    (stats['quantity_sold'] > stats['quantity_sold'].mean()) & 
    (stats['Revenue'] < stats['Revenue'].mean())
]

print("Potential Loss Leaders / High Volume Categories:")
print(high_vol_low_rev)
```

```python
df['day_of_week'] = df['transaction_date'].dt.day_name()

day_analysis = df.groupby('day_of_week')['Revenue'].mean().sort_values(ascending=False)
print("Highest Earning Days:\n", day_analysis)
```

```python
# 6. Which products frequently face low inventory but high demand?
```

```python
low_inventory_df = df[df['inventory_level'] <= df['reorder_point']]

product_risk = (
    low_inventory_df
    .groupby('product_name')
    .agg(
        total_demand=('actual_demand', 'sum'),
        avg_inventory=('inventory_level', 'mean'),
        stockout_count=('stockout_indicator', 'sum')
    )
    .reset_index()
)

critical_products = product_risk.sort_values(
    by=['total_demand', 'stockout_count'],
    ascending=[False, False]
)
```

```python
top_products = critical_products.head(10)

plt.figure(figsize=(10,6))
plt.plot(top_products['product_name'], top_products['total_demand'])

plt.title('Products with High Demand but Low Inventory')
plt.xlabel('Total Demand')
plt.ylabel('Product Name')

plt.tight_layout()
plt.show()
```

```python
# 7. Are there stores that are overstocked? 
```

```python
overstock_df = df[df['inventory_level'] > 1.5 * df['forecasted_demand']]
```

```python
overstock_df = df[df['inventory_level'] > 1.5 * df['forecasted_demand']]
store_overstock = (
    overstock_df
    .groupby(['store_id', 'store_location'])
    .agg(
        avg_inventory=('inventory_level', 'mean'),
        avg_forecasted_demand=('forecasted_demand', 'mean'),
        overstock_count=('inventory_level', 'count')
    )
    .reset_index()
)
store_overstock = store_overstock.sort_values(by='overstock_count', ascending=False)
top_stores = store_overstock.head(10)
plt.figure(figsize=(12,6))

x = np.arange(len(top_stores['store_location']))
width = 0.35

plt.bar(x - width/2, top_stores['avg_inventory'], width, label='Avg Inventory', color='skyblue')
plt.bar(x + width/2, top_stores['avg_forecasted_demand'], width, label='Avg Forecasted Demand', color='orange')

plt.title('Top 10 Overstocked Stores: Inventory vs Forecasted Demand')
plt.xlabel('Store Location')
plt.ylabel('Units')
plt.xticks(x, top_stores['store_location'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 9. Products with slow movement but high inventory. 
```

```python
avg_demand = df['actual_demand'].mean()
avg_inventory = df['inventory_level'].mean()
slow_moving = df[
    (df['actual_demand'] < avg_demand) &
    (df['inventory_level'] > avg_inventory)
]
```

```python
# 10. How inventory issues affect sales.
sales_stockout = (
    df.groupby('stockout_indicator')
    .agg(
        total_sales=('actual_demand', 'sum'),
        total_revenue=('Revenue', 'sum'),
        avg_inventory=('inventory_level', 'mean')
    )
    .reset_index()
)

print(sales_stockout)
```

```python
revenue_loss = (
    sales_stockout.loc[sales_stockout['stockout_indicator'] == 0, 'total_revenue'].values[0]
    -
    sales_stockout.loc[sales_stockout['stockout_indicator'] == 1, 'total_revenue'].values[0]
)

print(f"Estimated Revenue Impact Due to Inventory Issues: ${revenue_loss:,.2f}")
```

***.Forecast error = |forecasted âˆ’ actual demand|***

```python
#11.Forecast error = |forecasted âˆ’ actual demand| 
df['forecast_error'] = abs(df['forecasted_demand'] - df['actual_demand'])
print(df[['forecasted_demand', 'actual_demand', 'forecast_error']].head())
```

***Which categories have the worst forecasting accuracy?***

```python
# 12. Which categories have the worst forecasting accuracy?
df['forecast_error'] = abs(df['forecasted_demand'] - df['actual_demand'])
category_error = df.groupby('category')['forecast_error'].mean().reset_index()
worst_categories = category_error.sort_values(by='forecast_error', ascending=False)
print(worst_categories.head(5))
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(
    data=worst_categories.head(5), 
    x='category', 
    y='forecast_error', 
    palette='Reds_r'
)
plt.title('Top 5 Categories with Worst Forecasting Accuracy', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Forecast Error', fontsize=12)
plt.show()
```

***Impact of under-forecasting (lost sales)***

```python
# 13. Impact of under-forecasting (lost sales).
df['under_forecast'] = df['forecasted_demand'] < df['actual_demand']
```

```python
df['lost_sales_qty'] = df['actual_demand'] - df['forecasted_demand']
df['lost_sales_qty'] = df['lost_sales_qty'].apply(lambda x: x if x > 0 else 0)
```

```python
df['lost_revenue'] = df['lost_sales_qty'] * df['unit_price']
```

```python
lost_by_category = df.groupby('category')['lost_revenue'].sum().reset_index()
lost_by_category = lost_by_category.sort_values(by='lost_revenue', ascending=False)

print(lost_by_category)
```

***Impact of over-forecasting (excess inventory).***

```python
# 14. Impact of over-forecasting (excess inventory).
df['over_forecast'] = df['forecasted_demand'] > df['actual_demand']
```

```python
# Excess quantity = forecasted demand - actual demand (only if over-forecasted)
df['excess_inventory_qty'] = df['forecasted_demand'] - df['actual_demand']
df['excess_inventory_qty'] = df['excess_inventory_qty'].apply(lambda x: x if x > 0 else 0)
```

```python
df['excess_inventory_cost'] = df['excess_inventory_qty'] * df['unit_price']
```

```python
# Example: excess inventory cost by category
excess_by_category = df.groupby('category')['excess_inventory_cost'].sum().reset_index()
excess_by_category = excess_by_category.sort_values(by='excess_inventory_cost', ascending=False)

print(excess_by_category)
```

```python
# Under-forecasting impact
df['lost_sales_qty'] = (df['actual_demand'] - df['forecasted_demand']).clip(lower=0)
df['lost_revenue'] = df['lost_sales_qty'] * df['unit_price']

# Lost revenue by category
lost_by_category = df.groupby('category')['lost_revenue'].sum().sort_values(ascending=False)
lost_by_category.plot(kind='bar', figsize=(10,6), color='red', title='Lost Revenue by Category')
```

***Average order value per customer.***

```python
total_revenue_per_customer = df.groupby('customer_id')['Revenue'].sum()
orders_per_customer = df.groupby('customer_id')['transaction_id'].count()

aov_per_customer = (total_revenue_per_customer / orders_per_customer).reset_index()
aov_per_customer.rename(columns={0:'AOV'}, inplace=True)

# Display result
print(aov_per_customer.head())
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=aov_per_customer.sort_values('AOV', ascending=False).head(10),
            x='customer_id', y='AOV', palette='Blues_r')
plt.title('Top 10 Customers by Average Order Value')
plt.xlabel('Customer ID')
plt.ylabel('Average Order Value')
plt.show()
```

***Repeat vs one-time buyers (if possible).***

```python
# 18. Repeat vs one-time buyers (if possible).
orders_per_customer = df.groupby('customer_id')['transaction_id'].count().reset_index()
orders_per_customer.rename(columns={'transaction_id': 'order_count'}, inplace=True)

orders_per_customer['buyer_type'] = orders_per_customer['order_count'].apply(
    lambda x: 'Repeat Buyer' if x > 1 else 'One-time Buyer'
)

print(orders_per_customer['buyer_type'].value_counts())
df = df.merge(orders_per_customer[['customer_id', 'buyer_type']], on='customer_id', how='left')
```

```python
# Sum quantity sold per category
category_demand = df.groupby('category')['quantity_sold'].sum().reset_index()

# Sort categories by demand descending
category_demand = category_demand.sort_values(by='quantity_sold', ascending=False)

# Check the result
print(category_demand)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=category_demand, x='category', y='quantity_sold', palette='Blues_r')
plt.title('Category-wise Customer Demand')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=0)  # Rotate category names if needed
plt.show()
```

```python
category_revenue = df.groupby('category')['Revenue'].sum().reset_index()
category_revenue = category_revenue.sort_values(by='Revenue', ascending=False)
print(category_revenue)
```

```python
# Total revenue per store location
store_revenue = df.groupby('store_location')['Revenue'].sum().reset_index().sort_values(by='Revenue', ascending=False)

# Total quantity sold per store location
store_quantity = df.groupby('store_location')['quantity_sold'].sum().reset_index().sort_values(by='quantity_sold', ascending=False)

# Average Order Value (Revenue / #orders) per store
store_orders = df.groupby('store_location')['transaction_id'].count().reset_index().rename(columns={'transaction_id':'num_orders'})
store_aov = pd.merge(store_revenue, store_orders, on='store_location')
store_aov['AOV'] = store_aov['Revenue'] / store_aov['num_orders']

# Display results
print(store_revenue)
print(store_quantity)
print(store_aov[['store_location','AOV']])
```

```python
# Merge revenue, quantity, and AOV into one DataFrame
store_summary = pd.merge(store_revenue, store_quantity, on='store_location')
store_summary = pd.merge(store_summary, store_aov[['store_location','AOV']], on='store_location')

print(store_summary)
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

# Bar width
bar_width = 0.35
locations = range(len(store_summary))

# Plot Revenue bars
ax1.bar([x - bar_width/2 for x in locations], store_summary['Revenue'], width=bar_width, color='green', label='Revenue')

# Plot Quantity bars
ax1.bar([x + bar_width/2 for x in locations], store_summary['quantity_sold'], width=bar_width, color='blue', label='Quantity Sold')

ax1.set_xlabel('Store Location')
ax1.set_ylabel('Revenue / Quantity Sold')
ax1.set_xticks(locations)
ax1.set_xticklabels(store_summary['store_location'], rotation=45)
ax1.legend(loc='upper left')

# Plot AOV on secondary axis
ax2 = ax1.twinx()
ax2.plot(locations, store_summary['AOV'], color='orange', marker='o', linewidth=2, label='AOV')
ax2.set_ylabel('Average Order Value (AOV)')
ax2.legend(loc='upper right')

plt.title('Store-wise Revenue, Quantity Sold, and AOV')
plt.tight_layout()
plt.show()
```
