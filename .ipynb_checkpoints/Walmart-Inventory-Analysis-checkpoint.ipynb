***Task:-5 Which categories generate the highest revenue?***

```python
plt.figure(figsize=(10,5))
bars = plt.bar(category_revenue.index, category_revenue.values)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             yval,
             f'${yval:,.0f}',
             ha='center',
             va='bottom')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=0)
plt.show()
```

***Insight: Revenue by Category***

- Over-dependence on top categories may increase revenue risk.

**Business Recommendation:**

**Business Impact:**

***Task:-9 Which products frequently face low inventory but high demand?***

```python
low_inventory_df = df[df['inventory_level'] <= df['reorder_point']]

product_risk = (
    low_inventory_df
    .groupby('product_name')
    .agg(
        total_demand=('actual_demand', 'sum'),
        avg_inventory=('inventory_level', 'mean'),
        stockout_count=('stockout_indicator', 'sum')
    )
    .reset_index()
)

critical_products = product_risk.sort_values(
    by=['total_demand', 'stockout_count'],
    ascending=[False, False]
)
```

```python
top_products = critical_products.head(10)

plt.figure(figsize=(10,6))
plt.plot(top_products['product_name'], top_products['total_demand'])

plt.title('Products with High Demand but Low Inventory')
plt.xlabel('Total Demand')
plt.ylabel('Product Name')

plt.tight_layout()
plt.show()
```

### Insight: High-Demand, Low-Inventory Risk
- Around **3040% of products** are below their reorder point.
- Top 10 risky products contribute **~2535% of total demand**.
- Multiple stockout events indicate lost sales opportunities.

**Business Risk:**
High-demand products with low inventory pose a major revenue and customer satisfaction risk.

**Recommendation:**
Prioritize inventory replenishment and supplier optimization for critical products.

***Task:-10 Are there stores that are overstocked?***

```python
overstock_df = df[df['inventory_level'] > 1.5 * df['forecasted_demand']]
store_overstock = (
    overstock_df
    .groupby(['store_id', 'store_location'])
    .agg(
        avg_inventory=('inventory_level', 'mean'),
        avg_forecasted_demand=('forecasted_demand', 'mean'),
        overstock_count=('inventory_level', 'count')
    )
    .reset_index()
)
store_overstock = store_overstock.sort_values(by='overstock_count', ascending=False)
top_stores = store_overstock.head(10)
plt.figure(figsize=(12,6))

x = np.arange(len(top_stores['store_location']))
width = 0.35

plt.bar(x - width/2, top_stores['avg_inventory'], width, label='Avg Inventory', color='skyblue')
plt.bar(x + width/2, top_stores['avg_forecasted_demand'], width, label='Avg Forecasted Demand', color='orange')

plt.title('Top 10 Overstocked Stores: Inventory vs Forecasted Demand')
plt.xlabel('Store Location')
plt.ylabel('Units')
plt.xticks(x, top_stores['store_location'], rotation=45)
plt.legend()
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

### Insight: Store-Level Overstock Risk
- Nearly **2535% of stores** are overstocked.

- Overstocked stores hold **thousands of excess units**, increasing holding costs.

- Top 10 stores contribute **over 50% of overstock events**.

**Business Risk:**
Overstock ties up working capital and increases markdown and storage costs.

**Recommendation:**
Optimize store-level forecasting and rebalance inventory across locations.

***Task:-11 Inventory vs sales correlation.***

```python
corr_value = df['inventory_level'].corr(df['actual_demand'])
print("Inventory vs Sales Correlation:", corr_value)
```

```python
danger_zone = df[(df['inventory_level'] < df['inventory_level'].median()) & 
                 (df['actual_demand'] > df['actual_demand'].median())]

waste_zone = df[(df['inventory_level'] > df['inventory_level'].median()) & 
                (df['actual_demand'] < df['actual_demand'].median())]

print(f"Items in Danger Zone (Stockout Risk): {len(danger_zone)}")
print(f"Items in Waste Zone (Overstock Risk): {len(waste_zone)}")

avg_price = df['unit_price'].mean()
potential_loss = (danger_zone['actual_demand'] - danger_zone['inventory_level']).sum() * avg_price
print(f"Estimated Revenue at Risk: ${potential_loss:,.2f}")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x='inventory_level', y='revenue', data=df, 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title(f'Inventory vs Revenue (Correlation: {corr_value:.4f})', fontsize=14)
plt.xlabel('Inventory Level (Units in Stock)')
plt.ylabel('Revenue (USD)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**The Problem:***
#There is a near-zero correlation ($0.0032$) between stock levels and revenue.

***The Result:***
#The 'random' distribution of data points in this scatter plot confirms that our current inventory strategy is not data-driven.

***Financial Impact:***
#This lack of coordination is the root cause of our $-\$752,513$ lost revenue impact. We are consistently out of stock of high-revenue items while over-stocking low-revenue ones.

***Task:-16 Impact of under-forecasting (lost sales).***

```python
df['under_forecast'] = df['forecasted_demand'] < df['actual_demand']
```

```python
df['lost_sales_qty'] = df['actual_demand'] - df['forecasted_demand']
df['lost_sales_qty'] = df['lost_sales_qty'].apply(lambda x: x if x > 0 else 0)
```

```python
df['lost_revenue'] = df['lost_sales_qty'] * df['unit_price']
```

```python
lost_by_category = df.groupby('category')['lost_revenue'].sum().reset_index()
lost_by_category = lost_by_category.sort_values(by='lost_revenue', ascending=False)

print(lost_by_category)
```

 $359M revenue lost due to under-forecasting  major business impact

 Electronics alone caused $185M lost revenue  highest risk category

 Forecasting inaccuracies are concentrated in high-demand categories  inventory shortage risk

***Task:-17Impact of over-forecasting (excess inventory).***

```python
df['over_forecast'] = df['forecasted_demand'] > df['actual_demand']
```

```python
df['excess_inventory_qty'] = df['forecasted_demand'] - df['actual_demand']
df['excess_inventory_qty'] = df['excess_inventory_qty'].apply(lambda x: x if x > 0 else 0)
```

```python
df['excess_inventory_cost'] = df['excess_inventory_qty'] * df['unit_price']
```

```python
excess_by_category = df.groupby('category')['excess_inventory_cost'].sum().reset_index()
excess_by_category = excess_by_category.sort_values(by='excess_inventory_cost', ascending=False)

print(excess_by_category)
```

 Appliances caused $172,956,300 excess inventory cost  highest overstock risk

 Electronics caused $170,422,600 excess inventory cost  second highest impact

 Over $343M capital tied up in excess inventory  major working capital inefficiency

 Overstocking is concentrated in major categories  forecasting model inefficiency

***Task:-22 Store-location customer behavior differences.***

```python
store_revenue = df.groupby('store_location')['revenue'].sum().reset_index().sort_values(by='revenue', ascending=False)

store_quantity = df.groupby('store_location')['quantity_sold'].sum().reset_index().sort_values(by='quantity_sold', ascending=False)

store_orders = df.groupby('store_location')['transaction_id'].count().reset_index().rename(columns={'transaction_id':'num_orders'})
store_aov = pd.merge(store_revenue, store_orders, on='store_location')
store_aov['AOV'] = store_aov['revenue'] / store_aov['num_orders']

print(store_revenue)
print(store_quantity)
print(store_aov[['store_location','AOV']])
```

```python
store_summary = pd.merge(store_revenue, store_quantity, on='store_location')
store_summary = pd.merge(store_summary, store_aov[['store_location','AOV']], on='store_location')

print(store_summary)
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
bar_width = 0.35
locations = range(len(store_summary))

ax1.bar([x - bar_width/2 for x in locations], store_summary['revenue'], width=bar_width, color='green', label='revenue')

ax1.bar([x + bar_width/2 for x in locations], store_summary['quantity_sold'], width=bar_width, color='blue', label='Quantity Sold')

ax1.set_xlabel('Store Location')
ax1.set_ylabel('revenue / Quantity Sold')
ax1.set_xticks(locations)
ax1.set_xticklabels(store_summary['store_location'], rotation=45)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(locations, store_summary['AOV'], color='orange', marker='o', linewidth=2, label='AOV')
ax2.set_ylabel('Average Order Value (AOV)')
ax2.legend(loc='upper right')

plt.title('Store-wise Revenue, Quantity Sold, and AOV')
plt.tight_layout()
plt.show()
```

 Los Angeles generated $3.28M revenue  highest-performing store (21.87% of total revenue)

 Dallas generated $2.90M revenue  lowest-performing store

 Revenue gap is $374,369 (12.89%)  performance variation across stores

 Los Angeles sold highest quantity (3,168 units)  strongest demand location

 Los Angeles also has highest AOV ($3,156)  high-value customer base

 Dallas has lowest AOV ($2,910)  opportunity to increase customer spending
